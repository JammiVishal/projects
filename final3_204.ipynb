```python
import matplotlib.pyplot as plt
import random
```

```python
number = int(input("enter list length : "))

listQ=[]
listAC = []
listMC = []
listMR = []
listAR = []
listP = []

for i in range (number):
    listQ.append(random.randint(a = 0 , b  = 10))
    listAC.append(random.randint(a = 0 , b  = 500))
    listMC.append(random.randint(a = 0 , b  = 100))
    listMR.append(random.randint(a = 0 , b  = 170))
    listAR.append(random.randint(a = 0 , b  = 150))
    listP.append(random.randint(a = 0 , b  = 100))
```

```python

x = listQ
y = listP
y1 = listAR
y2 = listMR
y3 = listMC
y4 = listAC 

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)

plt.show()

```

```python
listQ=[]
number = int(input("enter list length"))
print("Enter Quantity values")
for i in range (number):
    data = int(input())
    listQ.append(data)

listP=[]
number = int(input("enter list length"))
print("Enter Price values")
for i in range (number):
    data = int(input())
    listP.append(data)

listAR=[]
number = int(input("enter list length"))
print("Enter AR values")
for i in range (number):
    data = int(input())
    listAR.append(data)

listMR=[]
number = int(input("enter list length"))
print("Enter MR values")
for i in range (number):
    data = int(input())
    listMR.append(data)

listMC=[]
number = int(input("enter list length"))
print("Enter MC values")
for i in range (number):
    data = int(input())
    listMC.append(data)

listAC=[]
number = int(input("enter list length"))
print("Enter AC values")
for i in range (number):
    data = int(input())
    listAC.append(data)

import matplotlib.pyplot as plt

x = listQ
y = listP
y1 = listAR
y2 = listMR
y3 = listMC
y4 = listAC 
plt.plot(x,y,label="Demand")
plt.plot(x,y1,label="AR")
plt.plot(x,y2,label="MR")
plt.plot(x,y3,label="MC")
plt.plot(x,y4,label="AC")
plt.ylabel("price")
plt.xlabel("quantity")
plt.title("market structure")
plt.legend()
fig = plt.figure(figsize = (20,15))
plt.show()
```
